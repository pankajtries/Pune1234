```python
df.isnull().sum()
```

```python
df['raisedhands']=df['raisedhands'].fillna(df['raisedhands'].mean)
df.isnull().sum()
```

```python
df['Discussion']=df['Discussion'].fillna(df['Discussion'].mean)
df.isnull().sum()
```

```python
df['VisITedResources']=df['VisITedResources'].fillna(method='ffill')
df.isnull().sum()
```

```python
df['AnnouncementsView']=df['AnnouncementsView'].fillna(method='ffill')
df.isnull().sum()
```

```python
df['ParentAnsweringSurvey']=df['ParentAnsweringSurvey'].astype(bool)
df.sample(5)
```

```python
df['Discussion'] = pd.to_numeric(df['Discussion'], errors='coerce')
print("Skewness before transformation:", df['Discussion'].skew())
df['Discussion_sqrt'] = np.sqrt(df['Discussion'])

print("Skewness after transformation:", df['Discussion_sqrt'].skew())
```

```python
df['raisedhands'] = pd.to_numeric(df['raisedhands'], errors='coerce')

print("Skewness before transformation:")
print(df['raisedhands'].skew())

df['raisedhands_sqrt'] = np.sqrt(df['raisedhands'])

print("Skewness after transformation:")
print(df['raisedhands_sqrt'].skew())
```

```python
df['AnnouncementsView'] = pd.to_numeric(df['AnnouncementsView'], errors='coerce')

print("Skewness before transformation:")
print(df['AnnouncementsView'].skew())

df['AnnouncementsView_sqrt'] = np.sqrt(df['AnnouncementsView'])

print("Skewness after transformation:")
print(df['AnnouncementsView_sqrt'].skew())
```
